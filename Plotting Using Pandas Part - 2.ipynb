```python
'''
1. figsize(length,width) increases or decreases the size of graph image.
2. orientation changes the axis origin I.e replaces x with y axis , y with x axis. parameter orientation='horizontal'
3. iloc [] is the index location of data .
'''

info2.iloc[8:17].plot(kind = 'bar',figsize=(15,10))
```

```python
info2.iloc[8:15].plot(kind = 'barh',figsize=(13,6)) # barh plots horizontal graph
```

```python
''' for single column data we pass the column enclosed in [] inside 'inverted comma '.
if we wish to get the data within a range we pass the range [a:b].  '''

info2['sepal_length'][5:12] 
```

```python
# when parameter vert = False changes the axis origin i.e replaces x with y axis ,y with x axis. 
info2.plot(kind = 'box',figsize=(15,6) , vert = False )
```

```python
info2.hist(figsize=(12,5), color='c', alpha = 0.5 )
```

```python
info2.plot(kind = 'area',figsize=(16,8))
```

```python
info2.plot(kind = 'area',figsize=(16,8) , stacked = False)
```

```python
# we use scatterplot to see the relation between 2 columns taken as parameter in x and y variable .
info2.plot.scatter(figsize=(9,6), x='sepal_length', y ='petal_length' )
```

```python
# here we used " s " as size parameter change the dot size whe ranges from 0 to n . 
info2.plot.scatter(figsize=(8,6), x='sepal_width', y ='petal_width', s=300 ,color ='y' , alpha = 0.4)
```

```python
info2.plot.scatter(figsize=(8,6), x='sepal_length', y ='petal_length', s= info2['petal_width']*100 ,color ='b',alpha = 0.5)
```

```python
# we use C 
info2.plot.hexbin( x='sepal_length', y ='petal_length'  ,figsize=(12,6) , gridsize = 20 , C='sepal_width' )
```
